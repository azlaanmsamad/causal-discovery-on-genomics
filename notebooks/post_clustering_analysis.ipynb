```python
true_cluster = pd.read_excel("data/1-s2.0-S0092867422005979-mmc3_table3.xlsx", sheet_name="gene expression clusters", index_col=0)
true_cluster
```

```python
cluster_member_dict = true_cluster['members'].to_dict()
cluster_member_dict
```

```python
cluster_member_dict = {k: v.split(",") for k,v in cluster_member_dict.items()}
#cluster_member_dict
```

```python
def has_intersection_over_alpha(set1, set2, alpha, base="union"):
    """
    Checks if the intersection of two sets is greater than alpha
    (a float between 0.0 and 1.0) of a base set size.

    The "Jaccard Index" is used when base="union".

    Args:
        set1 (set): The first set.
        set2 (set): The second set.
        alpha (float): The threshold (between 0.0 and 1.0).
        base (str, optional): The base for the percentage calculation.
            - "union": Jaccard similarity (intersection / union).
            - "smaller": intersection / size of the smaller set.
            - "larger": intersection / size of the larger set.
            - "set1": intersection / size of set1.
            - "set2": intersection / size of set2.
            Defaults to "union".

    Returns:
        bool: True if intersection percentage is > alpha, False otherwise.
        
    Raises:
        ValueError: If alpha is not between 0.0 and 1.0.
        ValueError: If the base string is not a valid option.
    """
    if not (0.0 <= alpha <= 1.0):
        raise ValueError("alpha must be a float between 0.0 and 1.0")

    intersection_size = len(set1.intersection(set2))
    len1 = len(set1)
    len2 = len(set2)

    base_size = 0
    if base == "union":
        base_size = len(set1.union(set2))
        # Per Jaccard Index definition, similarity of two empty sets is 1.0
        if base_size == 0:
            return 1.0 > alpha
    elif base == "smaller":
        base_size = min(len1, len2)
    elif base == "larger":
        base_size = max(len1, len2)
    elif base == "set1":
        base_size = len1
    elif base == "set2":
        base_size = len2
    else:
        raise ValueError("base must be one of 'union', 'smaller', 'larger', 'set1', 'set2'")

    # If the base size is 0 (e.g., base="set1" and set1 is empty),
    # the intersection is also 0. We define this similarity as 0.0.
    if base_size == 0:
        return False

    similarity = intersection_size / base_size
    return similarity > alpha
```

```python
mapping_true_to_pred = {}
for pred_cluster_num,pred_mem in predicted_cluster_member_dict.items():
    for true_cluster_num,true_mem in cluster_member_dict.items():
        if has_intersection_over_alpha(set(pred_mem), set(true_mem), alpha=0.75, base="set2"):
            if true_cluster_num in mapping_true_to_pred:
                mapping_true_to_pred[true_cluster_num].append(pred_cluster_num)
            else:
                mapping_true_to_pred[true_cluster_num] = [pred_cluster_num]
```

```python
for t,p in mapping_true_to_pred.items():
    print(f"----For True cluster Number {t}")
    print(f"true members: {cluster_member_dict[t]}")
    if len(p) ==1:
        print(f"Predicted member: {predicted_cluster_member_dict[p[0]]}")
    print(f"Common Members: {set(cluster_member_dict[t]).intersection(set(predicted_cluster_member_dict[p[0]]))}")
    print(f"\n")
```
